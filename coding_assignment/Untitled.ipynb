```python
price_dict = {'A': 100, 'B': 200, 'C': 300}
unit_dict = {'A': 1, 'B': 0, 'C': 5}


for key in price_dict:
        print(key)
        print("price: %s" % price_dict[key])
        print("unit: %s" % unit_dict[key])

    
for key in price_dict:
        value = price_dict[key] * unit_dict[key]
        print(value)
         
```
